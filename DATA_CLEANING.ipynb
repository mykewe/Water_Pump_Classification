### Data Cleaning

```python
# Extract numeric and categorical columns
numerical_cols = [cname for cname in train.columns if train[cname].dtype in ['int64', 'float64']]
categorical_cols = [cname for cname in train.columns if train[cname].dtype == "object" ]

print('numeric columns:', numerical_cols)
print('Categorical columns:', categorical_cols)
```

```python
# Get number of unique entries in each column with categorical data
object_nunique = list(map(lambda col: train[col].nunique(), categorical_cols))
d = dict(zip(categorical_cols, object_nunique))

# Print number of unique entries by column, in ascending order
sorted(d.items(), key=lambda x: x[1])
```

recorded_by has just one category and will be discarded

lga, date_recorded, funder, ward, installer, scheme_name, subvillage and wpt_name have too many categories ( greater than 21), I will first reduce these categories to make analysis easier

```python
# remove recorded_by
train.drop('recorded_by', axis=1, inplace=True);
all_data.drop('recorded_by', axis=1, inplace=True);
```

```python
cat_reduce = [cname for cname in train.columns if train[cname].dtype == "object" and train[cname].nunique() >21 ]
```

```python
def reduce_categories(data, col):
    a = data[col].value_counts()
    #get top 20 values of index
    vals = a[:20].index
    data[col+'_new'] = data[col].where(data[col].isin(vals), 'other')
```

```python
for col in cat_reduce:
    reduce_categories(train, col)
    reduce_categories(all_data, col)
```

```python
# remove the columns with many categories
train.drop(cat_reduce, axis=1, inplace=True);
all_data.drop(cat_reduce, axis=1, inplace=True);
```

```python
def percent(col):
    return (train[col].value_counts()/train[col].value_counts().sum()).round(4)* 100
```

```python
for col in numerical_cols:
    print(percent(col))
```

* amount_tsh is the amount of available water and it can't be zero but has 70% of zero values
* population also can't be be zero or 1
* construction_year also can't be 0 but has 34.9% of zero values

I will have to replace the zeros in this columns with the mean or median

```python
def replace_zero(data, col, method='mean' ):
    mtd = data[col].mean(skipna=True)
    if method=='median':
        mtd = data[col].median(skipna=True)
    data[col]=data[col].mask(data[col] == 0,mtd)
```

```python
def replace_one(data, col ):
    mtd = data[col].mean(skipna=True)
    data[col]=data[col].mask(data[col] == 1,mtd)
```

```python
zero_col = ['amount_tsh', 'population', 'construction_year']
```

```python
#I use mean for amount_tsh and population and median for construction year
replace_zero(train, 'amount_tsh')
replace_zero(all_data, 'amount_tsh')
replace_zero(train, 'population')
replace_zero(all_data, 'population')
replace_zero(train, 'construction_year', 'median')
replace_zero(all_data, 'construction_year', 'median')
```

```python
replace_one(train, 'population')
replace_one(all_data, 'population')
```

```python
for col in zero_col:
    print(percent(col))
```

### Null values

```python
train.isnull().mean().round(4) * 100
```

```python
na_cat_col = [ 'public_meeting', 'scheme_management', 'permit']
```

```python
#impute nulls for categorical data 
from sklearn.impute import SimpleImputer

for col in na_cat_col:
    imputer1 =SimpleImputer(strategy='most_frequent')
    imputer1.fit(train[[col]])
    train[col] = imputer1.transform(train[[col]]).ravel();
```

```python
for col in na_cat_col:
    imputer1 =SimpleImputer(strategy='most_frequent')
    imputer1.fit(all_data[[col]])
    all_data[col] = imputer1.transform(all_data[[col]]).ravel();
```

```python
#Save cleaned data
train.to_csv(path+'train_cleaned.csv')
all_data.to_csv(path+'all_data_cleaned.csv')
```
