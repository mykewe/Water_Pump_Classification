```python
# Import libraries and data
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
train = pd.read_csv("Training_data.csv", index_col='id')
label = pd.read_csv("Training_label.csv",index_col='id') 
test = pd.read_csv("Test_data.csv",index_col='id')
```

```python
Y=label.status_group.value_counts()
X=label.status_group.unique()
plt.bar(X,Y)
plt.show()
```

```python
train['label'] = label['status_group']
train['train_test'] = 1
test['train_test'] = 0
test['label'] = np.NaN
all_data = pd.concat([train,test]);
```

```python
# Extract numeric and categorical columns
numerical_cols = [cname for cname in train.columns if train[cname].dtype in ['int64', 'float64']]
categorical_cols = [cname for cname in train.columns if train[cname].dtype == "object" and train[cname].nunique() <20]

print('numeric columns:', numerical_cols)
print('Categorical columns:', categorical_cols)
```

```python
df_num = train[['amount_tsh', 'gps_height', 'longitude', 'latitude', 'num_private', 'region_code', 'district_code', 'population', 'construction_year']]
df_cat = train[['basin', 'public_meeting', 'recorded_by', 'scheme_management', 'permit', 'extraction_type', 'extraction_type_group', 'extraction_type_class', 'management', 'management_group', 'payment', 'payment_type', 'water_quality', 'quality_group', 'quantity', 'quantity_group', 'source', 'source_type', 'source_class', 'waterpoint_type', 'waterpoint_type_group']]
```

```python
#distributions for all numeric variables 
for i in df_num.columns:
    plt.hist(df_num[i])
    plt.title(i)
    plt.show() 
```
